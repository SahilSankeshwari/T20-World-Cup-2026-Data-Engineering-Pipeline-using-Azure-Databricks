```python
%sql
USE CATALOG t20_catalog;
USE SCHEMA silver;
```

**Create Clean Matches Table**

```python
df_matches = spark.table("t20_catalog.bronze.matches")

df_clean = df_matches.fillna("Unknown").dropDuplicates()

df_clean.write.format("delta") \
.mode("overwrite") \
.saveAsTable("t20_catalog.silver.matches_clean")
```

```python
spark.read.table("t20_catalog.silver.matches_clean").display()
```

**Clean Batting Stats**

```python
df = spark.table("t20_catalog.bronze.batting_stats")

df_clean = df.fillna({
    "runs": 0,
    "average": 0,
    "strike_rate": 0,
    "fours": 0,
    "sixes": 0,
    "hundreds": 0,
    "fifties": 0
}).dropDuplicates()

df_clean.write.format("delta") \
.mode("overwrite") \
.saveAsTable("t20_catalog.silver.batting_stats_clean")
```

```python
spark.read.table("t20_catalog.silver.batting_stats_clean").display()
```

**Clean Bowling Stats**

```python
df = spark.table("t20_catalog.bronze.bowling_stats")

df_clean = df.fillna({
    "wickets": 0,
    "runs_conceded": 0,
    "economy": 0,
    "average": 0
}).dropDuplicates()

df_clean.write.format("delta") \
.mode("overwrite") \
.saveAsTable("t20_catalog.silver.bowling_stats_clean")
```

```python
spark.read.table("t20_catalog.silver.bowling_stats_clean").display()
```

**Clean Points Table**

```python
df = spark.table("t20_catalog.bronze.points_table")

df_clean = df.fillna({
    "matches_played": 0,
    "won": 0,
    "lost": 0,
    "no_result": 0,
    "net_run_rate": 0,
    "points": 0
}).dropDuplicates()

df_clean.write.format("delta") \
.mode("overwrite") \
.saveAsTable("t20_catalog.silver.points_table_clean")
```

```python
spark.read.table("t20_catalog.silver.points_table_clean").display()
```
