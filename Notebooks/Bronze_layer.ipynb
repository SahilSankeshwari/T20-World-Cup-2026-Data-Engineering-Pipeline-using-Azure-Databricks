```python
%sql
USE CATALOG t20_catalog;
USE SCHEMA bronze;
```

**For matches**

```python
df_matches = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load("abfss://bronze@projectsto.dfs.core.windows.net/bronze/matches.csv")

display(df_matches)
```

```python
df_matches.write.format("delta") \
.mode("overwrite") \
.saveAsTable("t20_catalog.bronze.matches")
```
